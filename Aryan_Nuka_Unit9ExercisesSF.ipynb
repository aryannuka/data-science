Explain what happens when we do matrix multiplication.

Yes, I know I basically already asked you to do this in the notes. Don't copy and paste your anwser from before; honestly do the exercise again.

Each number in the resulting matrix is calculated by pairing a row from the first matrix with a column from the second. You multiply each element in the row by its corresponding element in the column, then add up all those products. This method is repeated for every combination of rows and columns.
